# Data Science Course Final Project: *Predicting Armed Conflict in the Middle East*

## 1. Problem Statement

For my final project, I worked to build a model to predict armed conlicts  in the middle east using the the Armed Conflict Location and Event Dataset (ACLED). The ability to understand the characteristics of events leading to violent clashes and predict violent clashes could potential save civilian and military lives. This information could be particularly useful for identifying the need for peacekeeping operations or for predicting increased terrorist activities in areas where the U.S. military is operating.

## 2. Approach Overview

### 2.1 Data Source
The Armed Conflict Location and Event Dataset (ACLED) contains records of conflicts, political violence and protests; including the dates, actors, types of violence, locations, and fatalities. The ACLED organization collects the dates, actors, types of violence, locations, and fatalities of all reported political violence and protest events around the world. Political violence and protest includes events that occur within civil wars and periods of instability, public protest and regime breakdown. ACLED captures the forms, actors, dates and locations of political violence and protest as it occurs. This data collection is funded through the Bureau of Conflict and Stabilization Operations (CSO) at the U.S. Department of State, the Dutch Ministry of Foreign Affairs, the Tableau Foundation, the International Organization for Migration (IOM), and the University of Texas at Austin.

The data used for this analysis consists of all events that occured between January 2016 and April 2020, in the Middle East. The primary limitation of this dataset is the limited timeframe, as there is just about 3.5 years of data. 

 Source: 
        “ACLED. (2017). “Armed Conflict Location & Event Data Project (ACLED) Codebook, 2017.””

### 2.2 Variables
ACLED contains 29 variables about each conflict recorded. For this analysis, I was interested in event fatalities, event type, sub-event type, actor(s) category, event location (country and administrative region) and date of event. Due to the large number of individual actors in the dataset, th general category of actors, recorded in INTER1 and INTER2, were used. Similarly, Event Type and Sub-Event Type were used to understand the type of conflict event. 

### 2.3 Model Approach
The problem described above is a classification problem - I am attempting to classify people into locations of seeking care. I tested three different classification models for this problem. The first was a KNearestNeighbors model, the second a decision tree model, and the third a random forest model. I will use prediction accuracy as the primary metric to evaluate each model. I may also look at Gini index as an evaluation metric.

## 5. More Data Manipulation
### 5.1 Dummy Code Variables

```python
ACLED_dummies = pd.get_dummies(ACLED, columns=['SUB_EVENT_TYPE', 'EVENT_TYPE',  'INTER1', 'INTER2'], drop_first = True)
ACLED_dummies.head()
```

### 5.2 Aggregate Data to Administrative Area by Date

```python
# Identify dummy columns that will need to be summed during aggregation
sum_cols = ACLED_dummies.drop(['ADMIN1','COUNTRY'],  axis=1).columns

# Create a Data Frame called locations with unique Administrative Area and Country Combination
locations = ACLED_dummies.groupby(['ADMIN1','COUNTRY'],axis='rows', 
                                  as_index=False
                                 ).count().loc[:,['ADMIN1', 'COUNTRY']].sort_values(by='COUNTRY')
print(locations.shape)
print(locations.head())
```

```python
# Loop through selecting subsets of ACLED_dummies by unique location
# Aggregated the subsetted dataset by 1 week using time index
# Reassign location information to the aggregated subset and append to ACLED_dummies_group

ACLED_dummies_group = pd.DataFrame()
for i in range(0,len(locations)):
    temp = ACLED_dummies.loc[
        (ACLED_dummies.loc[:,'ADMIN1'] == locations['ADMIN1'][i])
        & (ACLED_dummies.loc[:,'COUNTRY'] == locations['COUNTRY'][i])
        , sum_cols]
    temp = temp.resample('1W').sum()
    temp['COUNTRY'] = locations['COUNTRY'][i]
    temp['ADMIN1'] = locations['ADMIN1'][i]
    if len(ACLED_dummies_group) > 0:
        ACLED_dummies_group = ACLED_dummies_group.append(temp)
    else: 
        ACLED_dummies_group = temp
ACLED_dummies_group.head()
```

```python
# Get dummies for Country values to include this variable in the predication
ACLED_dummies_group['COUNTRY2'] = ACLED_dummies_group['COUNTRY'] # This will be used later for a timeshift and then dropped. 
ACLED_dummies_group = pd.get_dummies(ACLED_dummies_group, columns=['COUNTRY'], drop_first = True)
ACLED_dummies_group.head()
```

```python
# Shape of the final dataset
ACLED_dummies_group.shape
```

### 5.3 Timeshift data

#### Calculate Time-related variables

```python
# Use DATE field to calculate the month and assign to MONTH column
ACLED_dummies_group['DATE'] = ACLED_dummies_group.index
ACLED_dummies_group['MONTH'] = ACLED_dummies_group['DATE'].apply(lambda x:x.month)

# Determine Season based on Month and assign to SEASON column
def set_season(row):
    if row == 12 or row <= 2:
        return 'Winter'
    elif row >= 3 and row <= 5:
        return 'Spring'
    elif row >= 6 and row <= 8:
        return 'Summer'
    else:
        return 'Autumn'
    
ACLED_dummies_group['SEASON'] = ACLED_dummies_group['MONTH'].apply(set_season)
ACLED_dummies_group = pd.get_dummies(ACLED_dummies_group, columns=['SEASON','MONTH'], drop_first = True)
ACLED_dummies_group.head()
```

```python
# Identify columns that should be time shifted
drop_cols1 = ACLED_dummies_group.filter(like='SEASON', axis=1).columns
drop_cols2 = ACLED_dummies_group.filter(like='COUNTRY', axis=1).columns
time_feature_cols = ACLED_dummies_group.drop(['ADMIN1','DATE'],axis=1)
time_feature_cols = time_feature_cols.drop(drop_cols1,axis=1)
time_feature_cols = time_feature_cols.drop(drop_cols2,axis=1).columns

time_feature_cols
```

```python
temp = ACLED_dummies_group.loc[
    (ACLED_dummies_group.loc[:,'ADMIN1'] == locations['ADMIN1'][i])
    & (ACLED_dummies_group.loc[:,'COUNTRY2'] == locations['COUNTRY'][i])
    , sum_cols]
```

```python
# Includ month columns to sum_cols to retain after timeshift
sum_cols = list(sum_cols) + list(ACLED_dummies_group.filter(like='MONTH', axis=1).columns)
sum_cols
```

```python
# Calculate timeshift for individual location
ACLED_dummies_group_shift = pd.DataFrame()
for i in range(0,len(locations)):
    temp = ACLED_dummies_group.loc[
        (ACLED_dummies_group.loc[:,'ADMIN1'] == locations['ADMIN1'][i])
        & (ACLED_dummies_group.loc[:,'COUNTRY2'] == locations['COUNTRY'][i])
        , sum_cols]
 
    for feature in time_feature_cols:
        new_feature = feature + '_2wks_ago'
        temp.loc[:, new_feature] = temp.loc[:, feature].shift(1)
    # Delete first row containing null values
    temp = temp.iloc[1:,:]

    if len(ACLED_dummies_group_shift) > 0:
        ACLED_dummies_group_shift = ACLED_dummies_group_shift.append(temp)
    else: 
        ACLED_dummies_group_shift = temp
        
ACLED_dummies_group_shift.head()
```

```python
print(ACLED_dummies_group_shift.shape)
```

The final dataset for analysis has 31,828 rows and 112 columns.  This is a significant decrease from the 133K rows in the initial dataset and an increase in columns from the original 8 to 112. However, in model testing aggregation by date improved performance. 

## 5. Model Building 

### 5.1 Decision Tree Model

```python
#set feature cols as all columns of sub_acled except LOCATION DATE and FATALITIES
feature_cols = [feature + '_2wks_ago' for feature in time_feature_cols]
target_cols = 'FATALITIES'

# Set X and y
X = ACLED_dummies_group_shift.loc[:, feature_cols]
y = ACLED_dummies_group_shift.loc[:, 'FATALITIES']
```

```python
#Perform train_test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)
```

```python
# run preliminary model
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
predictions = linreg.predict(X_test)
lr_rmse = np.sqrt(mean_squared_error(y_test, predictions))
lr_rmse
```

The RMSE of the model is approximately 23.

```python
# asses model score
lin_train_score2 = linreg.score(X_train, y_train)
lin_test_score2 = linreg.score(X_test, y_test)
print(lin_train_score2,lin_test_score2 )
```

### 5.2 Random Forest

```python
# Instatiate decision tree model
lr = RandomForestRegressor
```

```python
# Instatiate decision tree model
X_rftrain, X_rftest, y_rftrain, y_rftest = train_test_split(X, y)
rf = RandomForestRegressor()

# Fit the model to my oversampled training set
rf.fit(X_rftrain, y_rftrain)

rf_train_score2 = linreg.score(X_rftrain, y_rftrain)
rf_test_score2 = linreg.score(X_rftest, y_rftest)
print(rf_train_score2,rf_test_score2 )
```

```python
# Use leave-one-out cross-validation (LOOCV) to estimate the RMSE for this model.
scores = cross_val_score(rf, X, y, cv=14, scoring='neg_mean_squared_error')
rf_rmse = np.mean(np.sqrt(-scores))
rf_rmse
```

The RMSE of the model is approximately 23.
